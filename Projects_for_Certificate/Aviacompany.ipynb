```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
df_train = pd.read_csv('./train_dataset.csv')
df_train.sample(5)
```

```python
df_train['satisfaction'].value_counts()
```

```python
df_train.info()
```

```python
df_train['Arrival Delay in Minutes'].isna().value_counts()
```

```python
df_train.dropna(inplace=True)
```

```python
df_train['Arrival Delay in Minutes'].isna().value_counts()
```

```python
df_train.drop('id', axis=1, inplace=True)
df_train.head()
```

```python
df_train['Class'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
encoder = OneHotEncoder()
columns_to_encode = ['Customer Type', 'Gender', 'Type of Travel', 'Class']
df=pd.DataFrame(encoder.fit_transform(df_train[columns_to_encode]).toarray(),columns=encoder.get_feature_names_out(), index=df_train.index)
```

```python
scaler = StandardScaler()
df_X = df_train.drop('satisfaction', axis=1)
cont_names = df_X.select_dtypes(include=['number']).columns.tolist()
scaler = StandardScaler()
df[cont_names]=pd.DataFrame(scaler.fit_transform(df_X[cont_names]),columns=cont_names, index=df_train.index)
```

```python
df.head()
```

```python
correlate_df = df.corr().abs()
correlate_df.style.background_gradient(cmap='coolwarm')
```

```python
df.corrwith(df_train['satisfaction']).abs().sort_values(ascending=False)
```

```python
X = df.values
y = df_train['satisfaction']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled[0]
df_train.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knc = KNeighborsClassifier(n_neighbors=5)
knc.fit(X_scaled, y)
```

```python
X_test = pd.read_csv('./test_dataset.csv')
X_test['Arrival Delay in Minutes'] = X_test['Arrival Delay in Minutes'].fillna(0.0)
X_test.drop('id', axis=1, inplace=True)
df_test=pd.DataFrame(encoder.transform(X_test[columns_to_encode]).toarray(),columns=encoder.get_feature_names_out(), index=X_test.index)
df_test[cont_names]=pd.DataFrame(scaler.fit_transform(X_test[cont_names]),columns=cont_names, index=X_test.index)

X_test_scaled = scaler.fit_transform(df_test)
df_test.head()
```

```python
y_test_predicted = knc.predict(X_test_scaled)
```

```python
ids = range(1, len(y_test_predicted) + 1)
rf_df = pd.DataFrame({
    'id' : ids,
    'satisfaction': y_test_predicted
})
rf_df.to_csv('sample_submission.csv', index=False)
rf_df
```

```python
from sklearn.model_selection import GridSearchCV
param = {'n_neighbors' : np.arange(1,25)}
grcv = GridSearchCV(knc, param, cv=5)
grcv.fit(X,y)
```
